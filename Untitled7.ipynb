```python
#task 06 
#Q.6: Draw Stripplot for each category against the following attributes ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary
#Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)','Calcium (% Daily Value)', 'Iron
#(% Daily Value)'. Here is one sample, you have to do it for all above mentioned attributes in list
```

```python
# Define the list of attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
              'Iron (% Daily Value)']
```

```python
# Create a function to plot Stripplot for each attribute
def plot_stripplot(attribute):
    plt.figure(figsize=(12, 8))
    sns.stripplot(x='Category', y=attribute, data=data, jitter=True)
    plt.xticks(rotation=45, ha='right')
    plt.title(f'{attribute} Distribution by Category')
    plt.show()
```

```python
# Plot Stripplot for each attribute
for attribute in attributes:
    plot_stripplot(attribute)
```

```python
#task 07 
#Draw a horizontal bar graph for items in each category against the calories.
```

```python
# Filter out unique categories
categories = data['Category'].unique()
```

```python
# Create a function to plot the horizontal bar graph for each category
def plot_category_bar(category):
    plt.figure(figsize=(12, 8))
    category_data = data[data['Category'] == category]
    category_data = category_data.sort_values(by='Calories', ascending=True)
    plt.barh(category_data['Item'], category_data['Calories'])
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.title(f'Calories in Each Item of {category}')
    plt.show()
```

```python

# Plot the horizontal bar graph for each category
for category in categories:
    plot_category_bar(category)
```
